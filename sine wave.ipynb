```python
%matplotlib inline
import numpy as np
import matplotlib
import matplotlib.pyplot as plt

## code showing 
## 1. how np.cos() returns a scaler or an array, depending on the input
## 2. scatter plot and regular plot with dots & lines
## 3. manipulating what to plot within the call

b = 2*np.pi*np.random.random_sample()
print(b)
c = np.cos(b)
print(type(c), c)

b = [2*np.pi*x for x in np.linspace(0, 1, 100)]
#print(type(b), b)

c = np.cos(b)
#print(type(c), c)

fig, ax = plt.subplots(2, 2, figsize=(15, 15))
ax[0][0].scatter(b, c)
ax[0][1].plot(b, c, '-o', color='red')
ax[1][0].plot(b, 'o', color='green')
# ax[1][1].plot([i-j for i, j in zip(b[1:], b[:-1])], '-o', color='green')
# even better:
ax[1][1].plot(np.diff(b), '-o', color='green')
plt.show()
```
