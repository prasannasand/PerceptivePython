```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

frequency = 440
num_samples = 48000

sampling_rate = 48000.0
amplitude = 16000

sine_wave = [amplitude * np.sin(2 * np.pi * frequency * x/num_samples) for x in range(num_samples)]
#
#sine_wave = amplitude * np.sin(2 * np.pi * frequency * np.linspace(0, num_samples) / sampling_rate)
```

```python
plt.plot(sine_wave[:500])
plt.show()
```

```python
noisy_sine = sine_wave + 1000 * np.random.randn(len(sine_wave))
plt.plot(noisy_sine[:500], color='red')
plt.plot(sine_wave[:500], color='green')
plt.show()
```

```python
fig, ax = plt.subplots(2, 2, figsize=(15, 15))
ax[0][0].scatter(b, c)
ax[0][1].plot(b, c, '-o', color='red')
ax[1][0].plot(b, 'o', color='green')
plt.show()
```
